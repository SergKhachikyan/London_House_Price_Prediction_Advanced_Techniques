```python
from xgboost import XGBRegressor
```

```python
X = df_train.drop('price', axis=1).copy()
y = df_train['price'].copy()
X_test = df_test.copy()

drop_cols = ['ID', 'fullAddress', 'date', 'address', 'number', 'street', 'paon', 'saon', 'town']
X = X.drop(columns=[col for col in drop_cols if col in X.columns])
X_test = X_test.drop(columns=[col for col in drop_cols if col in X_test.columns])

X = X.fillna(-1)
X_test = X_test.fillna(-1)

for col in X.columns:
    if X[col].dtype == 'object' or X_test[col].dtype == 'object':
        le = LabelEncoder()
        all_vals = pd.concat([X[col], X_test[col]]).astype(str)
        le.fit(all_vals)
        X[col] = le.transform(X[col].astype(str))
        X_test[col] = le.transform(X_test[col].astype(str))

model1 = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    tree_method='hist',
    random_state=42
)

model1.fit(X, y)
```

```python
import xgboost as xgb
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error
import numpy as np

kf = KFold(n_splits=5, shuffle=True, random_state=42)
val_scores = []

for train_idx, val_idx in kf.split(X):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval = xgb.DMatrix(X_val, label=y_val)

    params = {
        'objective': 'reg:squarederror',
        'learning_rate': 0.03,
        'max_depth': 6,
        'subsample': 0.8,
        'colsample_bytree': 0.8,
        'eval_metric': 'mae',
        'seed': 42
    }

    model3 = xgb.train(
        params,
        dtrain,
        num_boost_round=5000,
        evals=[(dval, 'eval')],
        early_stopping_rounds=100,
        verbose_eval=False
    )

    val_pred = model3.predict(dval)
    score = mean_absolute_error(y_val, val_pred)
    val_scores.append(score)

print(f"📉 CV MAE: {np.mean(val_scores):.0f}")
```

```python
dtest = xgb.DMatrix(X_test)
y_pred3 = model3.predict(dtest)
```
